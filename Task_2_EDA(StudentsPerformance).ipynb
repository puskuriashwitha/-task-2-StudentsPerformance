```python
# 1Ô∏è‚É£ Import Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 2Ô∏è‚É£ Load the Dataset
df = pd.read_csv("StudentsPerformance.csv")
print("Shape of the dataset:", df.shape)
print("\nFirst 5 rows:\n", df.head())

# 3Ô∏è‚É£ Check Data Types & Basic Info
print("\nData Types:\n")
print(df.dtypes)

print("\nDataset Info:\n")
print(df.info())

# 4Ô∏è‚É£ Check for Missing Values
print("\nMissing Values in Each Column:\n")
print(df.isnull().sum())

# 5Ô∏è‚É£ Check Unique Values / Value Counts (for Categorical Columns)
for col in df.select_dtypes(include='object').columns:
    print(f"\nValue counts for {col}:\n")
    print(df[col].value_counts())

# 6Ô∏è‚É£ Summary Statistics for Numeric Columns
print("\nSummary Statistics:\n")
print(df.describe())

# 7Ô∏è‚É£ Histograms for Score Columns
df[['Math_Score', 'Reading_Score', 'Writing_Score']].hist(figsize=(12, 6), bins=10)
plt.suptitle("Histograms of Scores")
plt.show()

# 8Ô∏è‚É£ Boxplots for Outliers Detection
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['Math_Score', 'Reading_Score', 'Writing_Score']])
plt.title("Boxplot for Scores")
plt.xticks(rotation=45)
plt.show()

# 9Ô∏è‚É£ Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# üîü Pairplot for Relationships
sns.pairplot(df[['Math_Score', 'Reading_Score', 'Writing_Score']])
plt.suptitle("Pairplot of Scores", y=1.02)
plt.show()
```
